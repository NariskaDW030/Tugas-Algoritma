```python
print ('Ayo masak hari ini !!')
masak = str (input ('Kamu mau makan apa ? (cah kangkung, tumis toge tahu, babi kecap) '))
if masak == 'cah kangkung' :
    print ('Kamu membutuhkan 1 ikat kangkung, 2 bawang putih dan 3 bawang merah, 1 cabai merah, dan 1 buah tomat')
if masak == 'tumis toge' :
    print ('Kamu membutuhkan 100 gram tauge kacang hijau, 250 gram tahu, 1 batang daun bawang dan seledri, cabe rawit, bawang merah dan putih secukupnya,dan bumbu secukupnya') 
if masak == 'babi kecap' :
    print ('Kamu membutuhkan 1 kg babi, kecap, bawang merah dan putih, 1 buah pala, 2 daun salam, l ruas jahe, 1 batang sereh')
dompet = float (input ('Lalu berapa uang yang kamu miliki didompet ?'))
if dompet >= 5000 and dompet < 15000 :
    print ('Kamu bisa membeli bahan untuk tumis toge tahu' )
if dompet > 15000 and dompet <= 25000 :
    print ('Kamu bisa membeli bahan untuk cah kangkung' )
if dompet > 25000 and dompet <= 50000 :
    print ('Kamu bisa membeli bahan untuk tumis toge dan cah kangkung' )
if dompet > 50000 :
    print ('Kamu bisa membeli bahan untuk babi kecap' )
elif dompet == 0 :
    print ('Sayang sekali kamu tidak bisa membeli apa-apa')
```
