```python
import numpy as numpi_30
import matplotlib.pyplot as plt

#Masukkan nilai a dan b
a_30 = numpi_30.array([0,1,2,3,4])
b_30 = numpi_30.array([0,1,4,9,16])
#Mencari nilai gradien dengan rumus y2-y1/x2-x1
#Disubstitusikan nilai a1_30,b1_30,a2_30,b2_30 ke y2-y1/x2-x1
m_30 = (b_30[-1]-b_30[0])/(a_30[-1]-a_30[0])
print ('Gradien dari dua titik tersebut adalah {}'.format(m_30))
print ('Persamaan garis melalui 2 titik adalah y={}x+{}'.format(m_30,-m_30*a_30+b_30))

#Membuat judul
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafik persamaan melalui 2 titik')
plt.legend()

#Menampilkan plot
plt.plot(a_30,b_30, label = '2x', color='red')
plt.xticks([0,1,2,3,4])
plt.yticks([0,1,4,9,16])
plt.show()
```

```python
import numpy as numpi_30
import matplotlib.pyplot as plt

#Membuat grafik jika diketahui 2 persamaan
#1.Membuat persamaan garis pertama jika diketahui titik M(a,b) dan N(c,d)
#Persamaan umum garis y=mx+b
a_30 = input('Masukkan titik a')
b_30 = input('Masukkan titik b')
c_30 = input('Masukkan titik c')
d_30 = input('Masukkan titik d')
#Lalu cari gradiennya dengan rumus y2-y1/x2-x1
#Substitusikan nilai a_30,b_30,c_30,d_30 ke y2-y1/x2-x1
#Menjadi m = d-b/c-a
Gradien_M_30 = int(d_30)-int(b_30)
Gradien_N_30 = int(c_30)-int(a_30)
Gradien1_30 = Gradien_M_30/Gradien_N_30
if Gradien1_30 == 0 :
    print ('Persamaan dari garis pertama tidak ada karena gradiennya 0, hanya akan menghasikan titik')
    #Jika gradien dari persamaan garis pertama 0, maka tidak ada persamaan dan hanya akan ada titik
else :
    print ('Gradien dari persamaan pertama adalah {}'.format(Gradien1_30))
    print ('Persamaan garis melalui 2 titik adalah y={}x+{}'.format(Gradien1_30,float(-Gradien1_30)*float(a_30)+float(b_30)))
    #Sehingga akan mendapatkan persamaan garis jika diketahui titik M dan N

#2.Membuat persamaan garis kedua jika diketahui titik L(h,i) dan O(j,k)
h_30 = input('Masukkan titik h')
i_30 = input('Masukkan titik i')
j_30 = input('Masukkan titik j')
k_30 = input('Masukkan titik k')
#Lalu cari gradiennya dengan menggunakan rumus sebelumnya
#Substitusikan nilai h_30,i_30,j_30,k_30 ke y2-y1/x2-x1
#Menjadi m = k-i/j-h
Gradien_L_30 = int(k_30)-int(i_30)
Gradien_O_30 = int(j_30)-int(h_30)
Gradien2_30 = Gradien_L_30/Gradien_O_30
print ('Gradien dari persamaan kedua adalah {}'.format(Gradien2_30))
print ('Persamaan garis melalui 2 titik adalah y={}x+{}'.format(Gradien2_30,float(-Gradien2_30)*float(h_30)+float(i_30)))
#Sehingga akan mendapatkan persamaan garis jika diketahui titik L dan O

#Kemudian menentukan titik potong dari kedua persamaan berikut, dan kedua gradien tidak boleh sama agar dapat ditentukan titik potongnya 
if Gradien1_30 == Gradien2_30 :
    print('Kedua gradien dari kedua persamaan garis sama, sehingga menyebabkan kedua garis tidak memiliki titik potong')
else :
    TPX_30 = (float(Gradien1_30)*float(a_30)-float(b_30)-float(Gradien2_30)*float(h_30)+float(i_30)/(float(Gradien1_30)-float(Gradien2_30)))
    TPY_30 = float(Gradien2_30)*(TPX_30-float(h_30))+float(i_30)
    print('Titik potong dari kedua persamaan garis tersebut adalah {},{}'.format(TPX_30,TPY_30))
    #Dan menghasilkan titik potong(x,y)

#Akan dibuat code untuk menampilakan grafik setelah diketahui persamaan kedua garis dan titik potongnya                                
x_30 =  numpi_30.arange(TPX_30-5,TPY_30+5,0.01)
plt.figure(figsize=(4,4),dpi=500)
plt.title('Grafik Persamaan Garis Lurus',fontdict = {'fontname': 'Times New Roman','fontsize':8})
plt.plot(x_30,Gradien1_30*(x_30-int(a_30))+int(b_30),color='green',label='y={}x+{}'.format(Gradien1_30,float(-Gradien1_30)*int(a_30)+int(b_30)))
plt.plot(x_30,Gradien2_30*(x_30-int(h_30))+int(i_30),color='blue',label='y={}x+{}'.format(Gradien2_30,float(-Gradien2_30)*int(h_30)+int(i_30)))
plt.plot(TPX_30,color='purple',marker='.',markersize= 4)
plt.plot(TPY_30,color='pink',marker ='.',markersize = 4)
plt.legend(loc='best')
plt.show()
```
