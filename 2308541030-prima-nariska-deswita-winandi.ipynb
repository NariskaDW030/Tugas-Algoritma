```python
#Jika diketahui bahwa ğ‘› adalah dua digit terakhir dari NIM kalian, maka tentukan biangan prima antara 1000ğ‘›Â danÂ 1000(ğ‘›+1).
#Diketahui NIM = 2308541030, diambil dua digit terakhir yaitu 30
#Maka 1000(30) dan 1000(30+1)
#Didapat 30000 dan 31000

print ('Jika n = 30, substitusikan nilai n ke 1000n dan 1000(n+1)')
def bil_prima_30(n_30):
    if n_30 <= 1:
        return False
    for i_30 in range(2, int(n_30**1/2) + 1):
        if n_30 % i_30 == 0:
            return False
    else:
        return True

#Substitusikan nilai n
n_30 = 30
prima_30 = []

#Tentukan rentang bilangan prima yang akan ditemukan yaitu antara 1000(ğ‘›) dan 1000(ğ‘›+1)
batas_bawah_30 = 1000 * n_30
batas_atas_30 = 1000 * (n_30 + 1)
print("Range bilangan prima yang akan dicari adalah {} dan {}\n".format(batas_bawah_30,batas_atas_30))

for angka_30 in range(batas_bawah_30, batas_atas_30):
    if bil_prima_30 (angka_30):
        prima_30.append(angka_30)
        
print("Daftar bilangan prima range 30000 dan 31000 adalah", prima_30)
#Hitung jumlah bilangan prima pada range 30000 - 31000
print("Jumlah bilangan prima dalam range tersebut adalah", len(prima_30))
     
```
