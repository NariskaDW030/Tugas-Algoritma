```python
# Persamaan Garis Jika diketahui 2 titik
print ('Mari bermain dengan titik-titik yang akan membentuk suatu garis')
import numpy as np
import matplotlib.pyplot as plt

# Masukkan titik garis pertama (a,b) dan (c,d)
n_30 = float(input("Nilai x dari titik pertama dari garis pertama adalah "))
a_30 = float(input("Nilai y dari titik pertama dari garis pertama adalah "))
r_30 = float(input("Nilai x dari titik kedua dari garis pertama adalah "))
i_30 = float(input("Nilai y dari titik kedua dari garis pertama adalah "))

# Masukkan titik garis kedua (e,f) dan (g,h)
s_30 = float(input("Nilai x dari titik pertama dari garis kedua adalah "))
k_30 = float(input("Nilai y dari titik pertama dari garis kedua adalah "))
d_30 = float(input("Nilai x dari titik kedua dari garis kedua adalah "))
w_30 = float(input("Nilai y dari titik kedua dari garis kedua adalah "))

# Gradien garis pertama
m1_30 = (i_30-a_30)/(r_30-n_30)
# Gradien garis kedua
m2_30 = (w_30-k_30)/(d_30-s_30)

# Persamaan garis pertama adalah y=m1(x-a)+b => m1*x-m1*a+b
# Persamaan garis kedua adalah y=m2(x-e)+f => m2*x-m2*e+f

# Titik potong garis pertama terhadap sumbu y
p1_30 = -m1_30*n_30+a_30
# Titik potong garis kedua terhadap sumbu y garis kedua
p2_30 = -m2_30*s_30+k_30
print("Persamaan garis pertama adalah y={}x+{}".format(m1_30,p1_30))
print("Persamaan garis kedua adalah y={}x+{}".format(m2_30,p2_30))

if m1_30== m2_30:
    #print("Kedua garis sejajar sehingga kemungkinan yang terjadi adalah tidak terdapat solusi atau terdapat tidak berhingga banyaknya solusi")
    if p1_30 == p2_30:
        print("Kedua garis akan berhimpitan sehingga terdapat tidak berhingga banyaknya solusi")
        x_30=np.linspace(-9,9,100)
        plt.plot(x_30,m1_30*(x_30-n_30)+a_30,color='red')
        plt.plot(x_30,m2_30*(x_30-s_30)+k_30,color='yellow')
        plt.title('Grafik persamaan melalui 2 titik')
        plt.show()
    else:
        print("Kedua garis tidak akan berpotongan, sehingga tidak akan terdapat solusi")
        x_30=np.linspace(-9,9,100)
        plt.plot(x_30,m1_30*(x_30-n_30)+a_30,color='blue')
        plt.plot(x_30,m2_30*(x_30-s_30)+k_30,color='green')
        plt.plot('Grafik persamaan melalui 2 titik')
        plt.show()
else:
    print("Akan terdapat satu titik potong")
    tp_x_30 = (-m2_30*s_30+k_30+m1_30*n_30-a_30)/(m1_30-m2_30)
    tp_y_30 = m1_30*(tp_x_30-n_30)+a_30
    print("Titik potong kedua garis yaitu ({},{})".format(tp_x_30,tp_y_30))
    x_30=np.linspace(tp_x_30-9,tp_x_30+9,100)
    plt.plot(x_30,m1_30*(x_30-n_30)+a_30,color='pink')
    plt.plot(x_30,m2_30*(x_30-s_30)+k_30,color='purple')
    plt.title('Grafik persamaan melalui 2 titik')
    plt.plot(tp_x_30,tp_y_30,color='green',marker='x',markersize=8)
    plt.axvline(x=0,color='black')
    plt.axhline(y=0,color='black')
    plt.show()
```

```python
print ('Mari kita buat grafik persamaan berbentuk ax + by = c')
import numpy as np
import matplotlib.pyplot as plt
# Masukkan nilai koefisien dan konstanta
print("Garis pertama yang memenuhi persamaan berbentuk ax + by = c ")
n_30 = float(input("Koefisien dari x pada garis pertama adalah "))
a_30 = float(input("Koefisien dari y pada garis pertama adalah "))
r_30 = float(input("Konstanta dari garis pertama adalah "))

print("Garis kedua yang memenuhi persamaan berbentuk dx + ey = f ")
i_30 = float(input("Koefisien dari x pada garis kedua adalah "))
s_30 = float(input("Koefisien dari y pada garis kedua adalah "))
k_30 = float(input("Konstanta dari garis kedua adalah "))
# Menghitung determinan utama
D_30 = n_30*s_30 - a_30*i_30

if D_30 == 0:
    print ("Garis akan berhimpitan, sehingga akan terdapat tak berhingga solusi.")
    print ("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi")
    x_vals = np.linspace(-10, 10, 400)
    y_vals1 = (r_30 - n_30*x_vals) / a_30
    y_vals2 = (k_30 - i_30*x_vals) / s_30

    # Plot garis pertama
    plt.plot(x_vals, y_vals1, label=f'{n_30}x + {a_30}y = {r_30}')

    # Plot garis kedua
    plt.plot(x_vals, y_vals2, label=f'{i_30}x + {s_30}y = {k_30}')

    # Plot titik potong
    plt.plot(x_30, y_30, 'go')  # 'go' means greeen color, circle marker
    plt.text(x_30, y_30, f'({x_30:.2f}, {y_30:.2f})', fontsize=14, ha='left')

    # Menambahkan label dan judul
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='brown',linewidth=0.5)
    plt.axvline(0, color='brown',linewidth=0.5)
    plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.title('Persamaan Garis ax+by=c')

    # Menampilkan plot
    plt.show()

else:
    # Menghitung determinan untuk x dan y
    Dx_30 = r_30*s_30 - a_30*k_30
    Dy_30 = n_30*k_30 - r_30*i_30
    # Menemukan solusi untuk x dan y
    x_30 = Dx_30 /D_30
    y_30 = Dy_30 /D_30
    print(f"Titik potong dari dua garis adalah: x = {x_30}, y = {y_30}")
    # Menyiapkan data untuk plotting
    x_vals = np.linspace(-10, 10, 400)
    y_vals1 = (r_30 - n_30*x_vals) / a_30
    y_vals2 = (k_30 - i_30*x_vals) / s_30

    # Plot garis pertama
    plt.plot(x_vals, y_vals1, label=f'{n_30}x + {a_30}y = {r_30}')

    # Plot garis kedua
    plt.plot(x_vals, y_vals2, label=f'{i_30}x + {s_30}y = {k_30}')

    # Plot titik potong
    plt.plot(x_30, y_30, 'go')  # 'go' means green color, circle marker
    plt.text(x_30, y_30, f'({x_30:.2f}, {y_30:.2f})', fontsize=14, ha='left')

    # Menambahkan label dan judul
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='brown',linewidth=0.5)
    plt.axvline(0, color='brown',linewidth=0.5)
    plt.grid(color = 'grey', linestyle ='--', linewidth = 0.5)
    plt.legend()
    plt.title('Persamaan Garis ax+by=c')
    plt.show()
    print('Tidak ada solusi ')
```
