```python
import numpy as np
import matplotlib.pyplot as plt
import math
 
print('Mari kita buat grafik persamaan ax^2+bx+c=0 dilengkapi dengan titik puncak')

a_30 = float (input ('Masukkan nilai a'))
b_30 = float (input ('Masukkan nilai b'))
c_30 = float (input ('Masukkan nilai c'))
# Masukkan nilai a,b,c ke dalam persamaan ax^2+bx+c

# Menghitung titik puncak dan sumbu simetri
def hitung_tpu_dan_ss(a_30, b_30, c_30):
    # Sumbu simetri (x) adalah -b / (2a)
    ss_30 = -b_30 / (2 * a_30)

    # Titik puncak (x, y)
    tpu_x_30 = ss_30
    tpu_y_30 = a_30 * (tpu_x_30 ** 2) + b_30 * tpu_x_30 + c_30

    return (tpu_x_30, tpu_y_30), ss_30

# Menentukan posisi titik potong sumbu y
def posisi_tp_y(c_30):
    if c_30 > 0:
        posisi = "potong sumbu y positif"
    elif c_30 < 0:
        posisi = "potong sumbu y negatif"
    else:
        posisi = "potong sumbu y di pusat (0,0)"
    return posisi

# Menentukan persamaan kuadrat
def plot_persamaan_kuadrat(a_30, b_30, c_30):
    # Menghitung titik puncak dan sumbu simetri
    tpu , ss_30 = hitung_tpu_dan_ss(a_30, b_30, c_30)

    # Menentukan posisi potong sumbu y
    tp_y = posisi_tp_y(c_30)

    # Membuat rentang x untuk plot
    x_30 = np.linspace(tpu[0] - 15, tpu[0] + 15, 400)
    y_30 = a_30 * x_30**2 + b_30 * x_30 + c_30

    # Rumus grafik persamaan kuadrat
    plt.figure(figsize=(10, 6))
    plt.plot(x_30, y_30, label=f'{a_30}x^2 + {b_30}x + {c_30}')

    # Titik potong sumbu y
    plt.plot(0, c_30, 'go')  
    plt.annotate(f'Potong sumbu y: {posisi_tp_y}\n(0, {c_30})',
                 xy=(0, c_30),
                 xytext=(-9, c_30 + 5),
                 arrowprops=dict(facecolor='pink', shrink=0.05))

    # Titik puncak
    plt.plot(tpu[0], tpu[1], 'ro') 
    plt.annotate(f'Titik puncak\n({tpu[0]:.2f}, {tpu[1]:.2f})',
                 xy=tpu,
                 xytext=(tpu[0] + 1, tpu[1] + 10),
                 arrowprops=dict(facecolor='black', shrink=0.05))

    # Sumbu simetri
    plt.axvline(x=ss_30, color='gray', linestyle='--', label='Sumbu simetri')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Grafik Persamaan Kuadrat dengan menunjukkan Titik Puncak')
    plt.axvline(x=0,color='black')
    plt.axhline(y=0,color='black')
    plt.legend()
    plt.grid(True)
    plt.show()

# Menjalankan fungsi plot_persamaan_kuadrat dengan nilai a, b, dan c yang diberikan
plot_persamaan_kuadrat(a_30, b_30, c_30)
```
